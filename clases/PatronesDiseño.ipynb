# Patrones de Diseño

## Introducción

### 1. ¿Qué son?

Los patrones de diseño son **soluciones habituales y probadas** para problemas que ocurren frecuentemente en el desarrollo de software.

* **No son código:** No son funciones ni bibliotecas que simplemente se copian y pegan.
* **Son conceptos:** Se trata de **planos prefabricados** o plantillas conceptuales.
* **Son adaptables:** Sos responsable de implementar la solución en tu código, adaptando los detalles para que encajen con la realidad específica de tu programa.

### 2. La Analogía del Plano (Blueprint)

Para entenderlos mejor, pensemos en un **plano de arquitectura**:

* El plano te muestra el resultado final (dónde van las paredes, las ventanas, la distribución).
* Sin embargo, vos decides qué materiales usar o el orden exacto de construcción según el terreno.
* Del mismo modo, un patrón te dice **cómo estructurar** la solución, pero la implementación final depende de vos.

### 3. Diferencia clave: Patrón vs. Algoritmo

Es común confundirlos porque ambos resuelven problemas, pero operan a niveles distintos.

| Característica | 🍳 Algoritmo (La Receta) | 📐 Patrón de Diseño (El Plano) |
| --- | --- | --- |
| **Definición** | Un conjunto claro de acciones paso a paso. | Una descripción de alto nivel de una solución. |
| **Analogía** | Una **receta de cocina**: sigues los pasos exactos para obtener el plato. | Un **plano de construcción**: ves el diseño y la función, pero tú construyes la estructura. |
| **Implementación** | Rígida. El objetivo es seguir el orden para lograr la meta. | Flexible. El orden y el código exacto dependen del programador. |
| **Resultado** | El mismo algoritmo suele verse igual en distintos programas. | El mismo patrón puede verse muy diferente en dos programas distintos. |



###  4.Estructura de un Patrón de Diseño

Para que un patrón sea fácil de reproducir en distintos contextos, se suele describir siguiendo un **formato estándar** y formal. Estas son las secciones clave que encontrarás en cualquier catálogo:

### A. Las 4 Partes Fundamentales

*  **Propósito:** Es el resumen ejecutivo. Explica brevemente el problema y la solución que propone el patrón.
*  **Motivación:** Profundiza en los detalles. Explica el contexto del problema y cómo la solución del patrón ayuda a resolverlo.
*  **Estructura de Clases:** Es la parte visual y técnica. Muestra cada una de las partes del patrón (clases) y cómo se relacionan entre sí.
*  **Ejemplo de Código:** Muestra la idea implementada en un lenguaje de programación popular (como Java, Python o C#) para facilitar la comprensión práctica.

### B. Detalles Adicionales

En catálogos más completos, también suelen incluirse:

* **Aplicabilidad:** ¿En qué situaciones específicas debo usarlo?
* **Pasos de implementación:** Guía paso a paso para integrarlo.
* **Relaciones:** Cómo interactúa o se diferencia de otros patrones.

## Historia y Origen

**1. ¿Invención o Descubrimiento?**
Los patrones no se inventan en un laboratorio, se **descubren**.

* Surgen cuando una solución a un problema común se repite una y otra vez en varios proyectos.
* Finalmente, alguien documenta esa solución repetida, le pone nombre y la explica en detalle. Así nace un patrón.

**2. La Inspiración: Arquitectura (Christopher Alexander)**
El concepto original no nació en la informática, sino en la **arquitectura**.

* Christopher Alexander escribió *El lenguaje de patrones*.
* Propuso un "lenguaje" para el diseño urbano (altura de ventanas, zonas verdes, niveles de edificios).
* Las unidades de este lenguaje eran los patrones.

**3. La "Banda de los Cuatro" (GoF)**
La idea saltó al software gracias a cuatro autores: *Erich Gamma, John Vlissides, Ralph Johnson y Richard Helm*.

* **Hito:** En **1995** publicaron el libro *"Patrones de diseño"*.
* **Contenido:** Adaptaron la idea de Alexander a la programación, presentando **23 patrones** para el diseño orientado a objetos.
* **El Nombre:** Debido a su título largo y a los cuatro autores, el libro y el grupo pasaron a conocerse mundialmente como la **Gang of Four (GoF)** o "La Banda de los Cuatro".

**4. Evolución Actual**
Aunque el libro GoF se centró en la Orientación a Objetos (POO), la metodología se expandió. Hoy existen patrones para muchas otras áreas (arquitectura de sistemas, UI/UX, programación funcional, etc.) más allá de los 23 originales.

## ¿Por qué aprender patrones?

Aunque es posible trabajar años sin estudiarlos formalmente (o incluso usarlos sin saberlo), aprenderlos conscientemente te da tres ventajas clave:

* **Caja de herramientas probada:** Son un conjunto de **soluciones comprobadas** a problemas comunes. Te evitan "reinventar la rueda" y te dan estrategias ya validadas.
* **Mejora tu diseño (POO):** Aunque no siempre uses un patrón específico, conocerlos te enseña a resolver problemas generales aplicando correctamente los principios de la **Orientación a Objetos**.
* **Vocabulario común:** Crean un lenguaje estándar para el equipo, haciendo la comunicación mucho más eficiente.
* *Ejemplo:* En lugar de explicar toda la lógica, basta con decir "Usa un Singleton" para que tus compañeros entiendan la estructura y la idea al instante.

## Crítica a los Patrones de Diseño
### Argumentos habituales contra el uso indiscriminado de los patrones:

- 1. Parche para lenguajes débiles

A menudo, la necesidad de un patrón surge porque el lenguaje de programación carece de cierto nivel de abstracción.

En estos casos, el patrón actúa como un "parche" para darle al lenguaje capacidades que no tiene nativamente.

Ejemplo: El patrón Strategy puede ser innecesario en lenguajes modernos que soportan funciones anónimas (lambdas).

- 2. Soluciones ineficientes (Dogmatismo)

Muchos desarrolladores ven los patrones como un dogma y los implementan "al pie de la letra".

Si no se adapta la solución al contexto específico del proyecto, se genera código ineficiente o innecesariamente complejo.

- 3. Uso injustificado (El "Martillo de Oro")

"Si lo único que tienes es un martillo, todo te parecerá un clavo".

Es un problema común en principiantes: una vez que aprenden los patrones, intentan forzarlos en todas partes, incluso en situaciones donde un código simple funcionaría mejor.

## Clasificación de los Patrones

Los patrones varían según su complejidad, nivel de detalle y escala. Se pueden organizar de dos formas principales:

**1. Por Nivel de Complejidad y Escala**

* **Idioms (Bajo nivel):**
* Son los patrones más básicos.
* Suelen ser específicos de un único lenguaje de programación.


* **Patrones de Arquitectura (Alto nivel):**
* Son universales y se pueden aplicar en casi cualquier lenguaje.
* Se utilizan para diseñar la estructura de una aplicación completa.
* *Analogía:* Es como comparar la instalación de un semáforo (solución simple) con la construcción de un intercambiador de autopistas (solución compleja).



**2. Por Propósito (Los 3 Grupos Principales)**

Todos los patrones se pueden clasificar según el objetivo que cumplen dentro del código:

 **Patrones Creacionales:**
* Se encargan de los mecanismos de **creación de objetos**.
* Objetivo: Incrementar la flexibilidad y la reutilización del código existente al crear instancias.


 **Patrones Estructurales:**
* Explican cómo **ensamblar objetos y clases** en estructuras más grandes.
* Objetivo: Mantener la estructura flexible y eficiente, garantizando que las partes encajen bien.


 **Patrones de Comportamiento:**
* Se encargan de la **comunicación efectiva** y la asignación de responsabilidades.
* Objetivo: Gestionar los algoritmos y la interacción entre objetos.

## Catálogo de Patrones de Diseño

Los patrones se dividen en tres familias principales según su propósito:

**1. Patrones Creacionales**

* **Enfoque:** Mecanismos de **creación de objetos**.
* **Objetivo:** Incrementar la flexibilidad y la reutilización del código existente, evitando las dependencias directas al instanciar clases.

**2. Patrones Estructurales**

* **Enfoque:** Cómo **ensamblar** objetos y clases.
* **Objetivo:** Formar estructuras más grandes y complejas (como puentes o adaptadores) manteniendo la eficiencia y la flexibilidad de la estructura resultante.

**3. Patrones de Comportamiento**

* **Enfoque:** **Comunicación** y asignación de responsabilidades.
* **Objetivo:** Gestionar algoritmos complejos y definir cómo interactúan y se reparten las tareas los distintos objetos entre sí.

## Listado de Patrones de Diseño

**1. Patrones Creacionales**
Se ocupan de la **creación de objetos** encapsulando la lógica de instanciación para desacoplar el sistema.

* **Factory Method:** Define una interfaz para crear un objeto, pero deja que las subclases decidan qué clase instanciar.
* **Abstract Factory:** Permite crear familias de objetos relacionados sin especificar sus clases concretas.
* **Builder:** Separa la construcción de un objeto complejo de su representación.
* **Prototype:** Crea nuevos objetos clonando una instancia existente (prototipo).
* **Singleton:** Garantiza que una clase tenga una única instancia y proporciona un punto de acceso global a ella.

**2. Patrones Estructurales**
Se ocupan de cómo se **componen las clases y objetos** para formar estructuras más grandes.

* **Adapter:** Permite que interfaces incompatibles trabajen juntas (como un adaptador de enchufe).
* **Bridge:** Desacopla una abstracción de su implementación para que puedan variar independientemente.
* **Composite:** Compone objetos en estructuras de árbol para representar jerarquías de "parte-todo".
* **Decorator:** Añade responsabilidades a un objeto de forma dinámica sin usar herencia.
* **Facade:** Proporciona una interfaz simplificada a un sistema complejo de clases o librerías.
* **Flyweight:** Comparte el estado para soportar un gran número de objetos de grano fino eficientemente (ahorro de RAM).
* **Proxy:** Proporciona un sustituto o intermediario para controlar el acceso a otro objeto.

**3. Patrones de Comportamiento**
Se ocupan de la **comunicación**, asignación de responsabilidades y algoritmos entre objetos.

* **Chain of Responsibility:** Pasa una solicitud a través de una cadena de manejadores.
* **Command:** Encapsula una petición como un objeto, permitiendo parametrizar clientes con diferentes solicitudes o colas.
* **Interpreter:** Define una representación gramatical para un lenguaje y un intérprete para interpretar oraciones.
* **Iterator:** Permite recorrer los elementos de una colección sin exponer su representación subyacente.
* **Mediator:** Define un objeto que encapsula cómo interactúan un conjunto de objetos (evita que se refieran entre sí explícitamente).
* **Memento:** Captura y restaura el estado interno de un objeto sin violar la encapsulación (para "deshacer").
* **Observer:** Define una dependencia uno-a-muchos; cuando un objeto cambia de estado, notifica a todos sus dependientes.
* **State:** Permite que un objeto altere su comportamiento cuando su estado interno cambia.
* **Strategy:** Define una familia de algoritmos, los encapsula y los hace intercambiables.
* **Template Method:** Define el esqueleto de un algoritmo en una operación, delegando algunos pasos a las subclases.
* **Visitor:** Separa un algoritmo de la estructura de objetos sobre la que opera.

No estudiaremos todos los patrones, nos centraremos en los siguientes:
- 1. Patrón Singleton (Creacional)
- 2. Patrón Repository (Estructural / Arquitectónico)
- 3. Patrón DTO (Data Transfer Object)

# Patrón Singleton - Resumen para Ciencia de Datos (Ciencia Política)

## ¿Qué es?

Singleton es un patrón de diseño que garantiza que una clase tenga **una única instancia** en todo el programa y proporciona un **punto de acceso global** a ella.

**Analogía política:** Piensen en el gobierno de un país. Sin importar cuántas veces mencionen "el Gobierno de Argentina", siempre se refieren a la misma entidad oficial. No puede haber dos gobiernos oficiales simultáneos.

## ¿Qué problema resuelve?

Cuando trabajamos con datos políticos, hay recursos que queremos controlar:

1. **Conexión a una base de datos** - No queremos abrir 50 conexiones diferentes a nuestra base de datos electoral
2. **Configuración del sistema** - Las configuraciones deben ser las mismas en todo el programa
3. **Registro de actividades (logs)** - Queremos un único archivo donde se registren todos los eventos

## ¿Cómo funciona?

**Dos características principales:**

1. **Constructor privado:** No se puede crear la instancia con `new` desde fuera de la clase
2. **Método estático:** Hay un método especial (generalmente `obtenerInstancia()`) que devuelve siempre el mismo objeto

## Ejemplo en pseudocódigo (simplificado)
```python
class ConexionBaseDatos:
    _instancia = None  # Variable privada para guardar LA instancia
    
    @staticmethod
    def obtener_instancia():
        if ConexionBaseDatos._instancia is None:
            ConexionBaseDatos._instancia = ConexionBaseDatos()
        return ConexionBaseDatos._instancia
    
    def consultar(self, sql):
        # Ejecutar consulta...
        pass

# Uso:
bd1 = ConexionBaseDatos.obtener_instancia()
bd1.consultar("SELECT * FROM elecciones")

bd2 = ConexionBaseDatos.obtener_instancia()
bd2.consultar("SELECT * FROM partidos")

# bd1 y bd2 son EL MISMO objeto
```

## Ejemplo aplicado a Ciencia Política

Imaginen que están analizando datos electorales y necesitan acceder a un archivo de configuración que contiene:
- Rutas a bases de datos
- Umbrales de corte electoral
- Parámetros de análisis

**SIN Singleton:** Cada parte del código crearía su propia copia de la configuración → Inconsistencias, desperdicio de memoria

**CON Singleton:** Todos acceden a la misma configuración → Coherencia garantizada

## Ventajas

✅ Garantiza una única instancia  
✅ Acceso global controlado  
✅ Se inicializa solo cuando se necesita (ahorro de recursos)

## Desventajas

❌ Viola el principio de responsabilidad única (hace dos cosas a la vez)  
❌ Puede dificultar las pruebas del código  
❌ Puede ocultar malas decisiones de diseño

## ¿Cuándo usarlo en análisis de datos políticos?

- Conexión a bases de datos electorales
- Gestor de configuraciones del proyecto
- Sistema de logging para auditoría
- Caché de datos censales que se consultan frecuentemente

## Conceptos clave para recordar

- **Una única instancia:** Como el Gobierno oficial de un país
- **Acceso global:** Todos pueden acceder, pero siempre al mismo objeto
- **Control centralizado:** Evita duplicación y garantiza consistencia

## Patrón Repository 

## ¿Qué es?

Repository es un patrón de diseño que **actúa como intermediario** entre la lógica de tu aplicación y la base de datos. Encapsula todas las operaciones de acceso a datos en un solo lugar.

**Analogía:** Pensá en un archivo central de un ministerio. Vos no vas directamente a los estantes a buscar los expedientes. Le pedís al archivista (el Repository) que te traiga lo que necesitás. El archivista sabe dónde está todo y cómo buscarlo, vos solo pedís y recibís.

## ¿Qué problema resuelve?

Cuando trabajamos con bases de datos enfrentamos estos desafíos:

1. **Código SQL disperso** - Consultas mezcladas por todo el código, difíciles de mantener
2. **Duplicación** - La misma consulta escrita varias veces en distintos lugares
3. **Acoplamiento** - Si cambias de SQL a MongoDB, tenés que modificar todo el código
4. **Testing difícil** - Es complicado testear lógica que tiene SQL embebido

## ¿Cómo funciona?

**Tres características principales:**

1. **Centralización:** Todas las operaciones de datos en una clase Repository
2. **Abstracción:** Tu código no sabe si los datos vienen de SQL, MongoDB o un CSV
3. **Interfaz simple:** Métodos con nombres claros como `obtener_diputado()` o `guardar_resultado_electoral()`

## Ejemplo en pseudocódigo (simplificado)
```python
class RepositorioElecciones:
    def __init__(self, db_connection):
        self.db = db_connection
    
    def obtener_resultado_por_provincia(self, provincia):
        # Acá va toda la lógica SQL compleja
        query = "SELECT * FROM resultados WHERE provincia = ?"
        return self.db.execute(query, [provincia])
    
    def obtener_resultados_historicos(self, año_desde, año_hasta):
        # Otra consulta compleja encapsulada
        query = """
            SELECT * FROM resultados 
            WHERE año BETWEEN ? AND ?
            ORDER BY año
        """
        return self.db.execute(query, [año_desde, año_hasta])
    
    def guardar_resultado(self, resultado):
        # Lógica de inserción encapsulada
        query = "INSERT INTO resultados VALUES (?, ?, ?)"
        self.db.execute(query, resultado.values())

# Uso desde tu aplicación:
repo = RepositorioElecciones(db_connection)

# Simple, claro, sin SQL en tu lógica
resultados_ba = repo.obtener_resultado_por_provincia("Buenos Aires")
historicos = repo.obtener_resultados_historicos(2015, 2023)
```

## Ejemplo aplicado a Ciencia Política

Imaginá que estás construyendo un sistema de análisis electoral:

**SIN Repository:**
```python
# En el archivo analisis.py
resultado = db.execute("SELECT * FROM resultados WHERE provincia = 'CABA'")

# En el archivo visualizacion.py  
resultado = db.execute("SELECT * FROM resultados WHERE provincia = 'CABA'")

# En el archivo exportar.py
resultado = db.execute("SELECT * FROM resultados WHERE provincia = 'CABA'")
```
→ Misma consulta repetida 3 veces, difícil de mantener

**CON Repository:**
```python
# En repository.py (un solo lugar)
class RepositorioElecciones:
    def obtener_por_provincia(self, provincia):
        return self.db.execute(
            "SELECT * FROM resultados WHERE provincia = ?", 
            [provincia]
        )

# En cualquier otro archivo
repo = RepositorioElecciones(db)
resultado = repo.obtener_por_provincia("CABA")
```
→ Consulta definida una vez, usada en todos lados

## Ventajas

✅ **Centralización:** Todo el código de datos en un solo lugar  
✅ **Mantenibilidad:** Cambiar una consulta no requiere buscar en 20 archivos  
✅ **Testeable:** Podés crear un "FakeRepository" para tests sin tocar la DB real  
✅ **Desacoplamiento:** Cambiar de SQL a MongoDB solo requiere cambiar el Repository  
✅ **Claridad:** Métodos con nombres descriptivos como `obtener_diputados_activos()`

## Desventajas

❌ Capa extra de abstracción (más archivos y clases)  
❌ Puede ser overkill para proyectos muy simples  
❌ Requiere disciplina para no saltear el Repository y hacer queries directas

## ¿Cuándo usarlo en análisis de datos políticos?

- Cuando tenés múltiples fuentes de datos (SQL + APIs + CSV)
- Si tu proyecto va a crecer y necesitás mantenibilidad
- Para proyectos colaborativos (cada uno sabe dónde está el código de datos)
- Cuando querés escribir tests unitarios sin depender de la DB real
- Si planeás migrar de base de datos en el futuro

## Conceptos clave para recordar

- **Intermediario:** Como el archivista del ministerio, media entre vos y los datos
- **Abstracción:** No te importa de dónde vienen los datos, solo cómo pedirlos
- **Single Responsibility:** Cada Repository maneja UNA entidad (Elecciones, Diputados, etc.)
- **No es una capa ORM:** Es más simple, es solo organización de código

# Patrón DTO (Data Transfer Object)
## ¿Qué es?

DTO es un patrón de diseño que consiste en **objetos simples que solo transportan datos** entre diferentes partes de tu aplicación o entre sistemas. No tienen lógica de negocio, solo datos.

**Analogía:** Pensá en un formulario oficial. Cuando tramitás un documento, llenás un formulario con tus datos (nombre, DNI, domicilio). El formulario no hace nada por sí mismo, solo organiza y transporta la información del mostrador A al mostrador B. Ese formulario es un DTO.

## ¿Qué problema resuelve?

Cuando construimos aplicaciones que consultan y envían datos, enfrentamos estos problemas:

1. **Exponer demasiado** - No querés enviar TODO lo que está en tu base de datos (ej: contraseñas, datos internos)
2. **Formato inadecuado** - La estructura de tu base de datos no es la misma que necesita tu frontend
3. **Acoplar capas** - Si el frontend usa directamente tus modelos de base de datos, cualquier cambio rompe todo
4. **Validación** - Necesitás asegurar que los datos que llegan cumplan ciertas reglas

## ¿Cómo funciona?

**Tres características principales:**

1. **Solo datos:** Sin lógica de negocio, solo propiedades/atributos
2. **Serializable:** Fácil de convertir a JSON para enviar por la red
3. **Específico para cada caso:** Un DTO diferente para cada "conversación" entre sistemas

## Ejemplo en pseudocódigo (simplificado)
```python
# Modelo de la Base de Datos (NO lo exponemos directamente)
class DiputadoDB:
    def __init__(self):
        self.id = 1
        self.nombre = "María González"
        self.dni = "12345678"
        self.partido_id = 5
        self.sueldo_bruto = 500000
        self.fecha_alta_sistema = "2023-01-15"
        self.password_hash = "asdfjkl234..."  # ¡Esto NUNCA debe salir!
        self.notas_internas = "..."

# DTO para enviar al frontend (solo lo necesario)
class DiputadoDTO:
    def __init__(self, nombre, partido, distrito):
        self.nombre = nombre
        self.partido = partido
        self.distrito = distrito
    
    def to_json(self):
        return {
            "nombre": self.nombre,
            "partido": self.partido,
            "distrito": self.distrito
        }

# Conversión (Mapping)
def convertir_a_dto(diputado_db, partido_nombre):
    return DiputadoDTO(
        nombre=diputado_db.nombre,
        partido=partido_nombre,
        distrito="CABA"
    )

# Uso en tu API
diputado_db = obtener_de_base_datos(id=1)
diputado_dto = convertir_a_dto(diputado_db, "PRO")
return diputado_dto.to_json()  # Solo datos seguros y necesarios
```

## Ejemplo aplicado a Ciencia Política

Imaginá que estás construyendo una API de datos electorales:

**CON DTO (BIEN - Solo lo necesario):**
```python
class ResultadoElectoralDTO:
    def __init__(self, provincia, resultados):
        self.provincia = provincia
        self.resultados = resultados
        self.total_votos = sum(resultados.values())
    
    def to_json(self):
        return {
            "provincia": self.provincia,
            "resultados": self.resultados,
            "total_votos": self.total_votos
        }

# ✅ Creás un DTO limpio
dto = ResultadoElectoralDTO(
    provincia="Buenos Aires",
    resultados={
        "Partido A": 1500000,
        "Partido B": 1200000,
        "En Blanco": 50000
    }
)
return dto.to_json()  # Solo datos relevantes para el usuario
```

## Ventajas

✅ **Seguridad:** No exponés campos sensibles (contraseñas, IPs, datos internos)  
✅ **Desacoplamiento:** Frontend y Backend pueden cambiar independientemente  
✅ **Optimización:** Enviás solo los datos necesarios (menos tráfico de red)  
✅ **Validación:** Podés validar formato y tipos antes de procesar  
✅ **Versionado:** DTOs diferentes para versiones diferentes de tu API (v1, v2)

## Desventajas

❌ Código extra (más clases para mantener)  
❌ Necesitás mapear entre modelos DB y DTOs (conversión manual o con librerías)  
❌ Puede ser overkill para proyectos muy simples  
❌ Duplicación aparente de estructuras de datos

## ¿Cuándo usarlo en análisis de datos políticos?

- Cuando construís APIs que exponen datos
- Si trabajás con datos sensibles que no podés mostrar todos
- Para separar la estructura interna de tu DB de lo que ve el usuario
- Cuando necesitás diferentes "vistas" de los mismos datos según el contexto
- Si tu frontend necesita datos en un formato específico diferente a tu DB

## Patrones comunes de DTOs

### 1. DTO de Entrada (Request)
```python
class CrearDiputadoDTO:
    def __init__(self, nombre, partido, distrito):
        self.nombre = nombre
        self.partido = partido
        self.distrito = distrito
    
    def validar(self):
        if not self.nombre:
            raise ValueError("Nombre es requerido")
        if len(self.nombre) < 3:
            raise ValueError("Nombre muy corto")
```

class DiputadoResumenDTO:

## DTO + Pydantic en FastAPI

En Python moderno, usamos **Pydantic** para crear DTOs con validación automática:
```python
from pydantic import BaseModel, Field, validator

class DiputadoCreateDTO(BaseModel):
    nombre: str = Field(..., min_length=3, max_length=100)
    partido: str
    distrito: str
    
    @validator('nombre')
    def nombre_debe_tener_mayuscula(cls, v):
        if not v[0].isupper():
            raise ValueError('Nombre debe empezar con mayúscula')
        return v

class DiputadoResponseDTO(BaseModel):
    id: int
    nombre: str
    partido: str
    distrito: str
    
    class Config:
        from_attributes = True  # Permite crear desde objetos DB
```

## Conceptos clave para recordar

- **Transportador de datos:** Como un formulario, solo lleva información
- **Sin lógica:** No procesa, no calcula, solo contiene datos
- **Capa de traducción:** Convierte entre el formato interno y el externo
- **Múltiples DTOs para un modelo:** Un modelo DB puede tener varios DTOs según el contexto
- **No es el modelo:** El DTO no es tu modelo de base de datos, es una representación para transferir

## DTO vs Modelo de Base de Datos

| Característica | Modelo DB | DTO |
|----------------|-----------|-----|
| **Propósito** | Representar estructura de la tabla | Transferir datos entre capas |
| **Lógica** | Puede tener métodos de negocio | Solo datos, sin lógica |
| **Campos** | Todos los campos de la tabla | Solo campos necesarios para esa operación |
| **Relaciones** | Relaciones con otras tablas | Puede incluir datos "aplanados" |
| **Validación** | Validaciones de DB (constraints) | Validaciones de entrada/salida |
| **Cambios** | Requiere migración de DB | Más flexible, no afecta DB |

**Pregunta para reflexionar:** Si tu base de datos de resultados electorales tiene 30 columnas pero tu mapa interactivo solo necesita mostrar 5, ¿conviene enviar las 30 o crear un DTO con las 5 necesarias?

**Consejo final:** En proyectos pequeños podés saltear DTOs, pero cuando construyas APIs que serán consumidas por otros (o por tu propio frontend), los DTOs son fundamentales para mantener un diseño limpio y seguro.
