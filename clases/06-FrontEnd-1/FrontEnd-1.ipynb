# Clase 7: HTML, CSS, Bootstrap y JavaScript - Fundamentos

## Introducción

En esta clase daremos nuestros primeros pasos en el **desarrollo frontend**: la parte visible de las aplicaciones web que los usuarios ven e interactan. 

Hasta ahora trabajamos con:
- Python (backend)
- Bases de datos (almacenamiento)
- APIs (comunicación)

Ahora vamos a construir **interfaces web** que consuman esas APIs y presenten los datos de forma atractiva e interactiva.

### El Stack Frontend Moderno
```
┌─────────────────────────────────────┐
│         HTML (Estructura)           │  ← El esqueleto
├─────────────────────────────────────┤
│         CSS (Presentación)          │  ← La piel y estilo
├─────────────────────────────────────┤
│      JavaScript (Comportamiento)    │  ← Los músculos y cerebro
├─────────────────────────────────────┤
│    Bootstrap (Framework CSS)        │  ← Herramientas pre-hechas
└─────────────────────────────────────┘
```

### Analogía política

Pensá en una **página de resultados electorales**:

- **HTML:** Define QUÉ elementos hay (tabla de resultados, gráfico de barras, mapa)
- **CSS:** Define CÓMO se ven (colores partidarios, tamaños, espaciado)
- **JavaScript:** Define la INTERACCIÓN (filtrar por provincia, actualizar en vivo, mostrar tooltips)
- **Bootstrap:** Componentes listos para usar (tablas responsive, cards, modales)

## 1. HTML5: La Estructura Semántica

### ¿Qué es HTML?

**HTML** (HyperText Markup Language) es el lenguaje de marcado que define la **estructura** de una página web. No es un lenguaje de programación, sino de **marcado**: describe el contenido usando etiquetas.

### Anatomía de una etiqueta HTML
```html
<etiqueta atributo="valor">Contenido</etiqueta>
│        │                 │           │
│        └─ Atributos      └─ Contenido └─ Cierre
└─ Apertura
```

### Estructura básica de un documento HTML5
```html
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Resultados Electorales 2023</title>
</head>
<body>
    <!-- Acá va todo el contenido visible -->
    <h1>Resultados Electorales</h1>
    <p>Última actualización: 22:00hs</p>
</body>
</html>
```

**Explicación línea por línea:**

- `<!DOCTYPE html>`: Declara que es HTML5
- `<html lang="es">`: Raíz del documento, idioma español
- `<head>`: Metadatos (no visibles al usuario)
  - `<meta charset="UTF-8">`: Codificación de caracteres (para tildes, ñ, etc.)
  - `<meta name="viewport" ...>`: Hace la página responsive (se adapta a móviles)
  - `<title>`: Título que aparece en la pestaña del navegador
- `<body>`: Todo el contenido visible

### HTML Semántico: Etiquetas con Significado

HTML5 introduce etiquetas **semánticas**: en lugar de usar `<div>` para todo, usamos etiquetas que describen el propósito del contenido.

#### ❌ HTML NO semántico (antiguo):
```html
<div id="header">
    <div id="nav">...</div>
</div>
<div id="main">
    <div class="article">...</div>
</div>
<div id="footer">...</div>
```

#### ✅ HTML Semántico (moderno):
```html
<header>
    <nav>...</nav>
</header>
<main>
    <article>...</article>
</main>
<footer>...</footer>
```

### Principales etiquetas semánticas

| Etiqueta | Propósito | Ejemplo de uso |
|----------|-----------|----------------|
| `<header>` | Encabezado de página o sección | Logo, título, menú principal |
| `<nav>` | Navegación principal | Menú de navegación |
| `<main>` | Contenido principal (único por página) | Cuerpo principal del artículo/app |
| `<article>` | Contenido independiente | Post de blog, noticia, resultado electoral |
| `<section>` | Sección temática | Agrupación de contenido relacionado |
| `<aside>` | Contenido relacionado pero secundario | Sidebar, info adicional |
| `<footer>` | Pie de página | Copyright, links, contacto |
| `<figure>` y `<figcaption>` | Imagen/gráfico con descripción | Gráfico electoral con título |

### Ejemplo: Página de Resultados Electorales
```html
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Resultados Electorales 2023 - CABA</title>
</head>
<body>
    <header>
        <h1>Resultados Electorales 2023</h1>
        <nav>
            <ul>
                <li><a href="#presidencial">Presidencial</a></li>
                <li><a href="#legislativas">Legislativas</a></li>
                <li><a href="#mapa">Mapa</a></li>
            </ul>
        </nav>
    </header>

    <main>
        <section id="presidencial">
            <h2>Elecciones Presidenciales</h2>
            
            <article class="candidato">
                <h3>Javier Milei - La Libertad Avanza</h3>
                <p>Votos: 5,500,000 (30.5%)</p>
            </article>

            <article class="candidato">
                <h3>Sergio Massa - Unión por la Patria</h3>
                <p>Votos: 5,200,000 (28.8%)</p>
            </article>
        </section>

        <aside>
            <h3>Datos del Escrutinio</h3>
            <p>Mesas escrutadas: 85%</p>
            <p>Última actualización: 22:30hs</p>
        </aside>
    </main>

    <footer>
        <p>&copy; 2023 - Tribunal Electoral Nacional</p>
    </footer>
</body>
</html>
```

### Etiquetas HTML más comunes

**Texto:**
- `<h1>` a `<h6>`: Títulos (h1 el más importante, h6 el menos)
- `<p>`: Párrafo
- `<strong>`: Texto importante (negrita semántica)
- `<em>`: Énfasis (cursiva semántica)
- `<span>`: Contenedor inline sin significado semántico

**Listas:**
```html
<!-- Lista desordenada -->
<ul>
    <li>Item 1</li>
    <li>Item 2</li>
</ul>

<!-- Lista ordenada -->
<ol>
    <li>Primero</li>
    <li>Segundo</li>
</ol>
```

**Enlaces e imágenes:**
```html
<!-- Enlace -->
<a href="https://ejemplo.com" target="_blank">Visitar sitio</a>

<!-- Imagen -->
<img src="grafico.png" alt="Gráfico de resultados electorales">
```

**Tablas (muy útiles para datos):**
```html
<table>
    <thead>
        <tr>
            <th>Partido</th>
            <th>Votos</th>
            <th>%</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>LLA</td>
            <td>5,500,000</td>
            <td>30.5%</td>
        </tr>
        <tr>
            <td>UxP</td>
            <td>5,200,000</td>
            <td>28.8%</td>
        </tr>
    </tbody>
</table>
```

### Accesibilidad en HTML

**¿Por qué importa?**
- Usuarios con discapacidades visuales usan lectores de pantalla
- Buscadores leen el HTML para indexar contenido
- Mejora la usabilidad para todos

**Buenas prácticas:**

1. **Usar atributo `alt` en imágenes:**
```html
<img src="mapa.png" alt="Mapa electoral de Argentina con resultados por provincia">
```

2. **Usar etiquetas `<label>` en formularios:**
```html
<label for="provincia">Provincia:</label>
<select id="provincia" name="provincia">
    <option value="caba">CABA</option>
    <option value="bsas">Buenos Aires</option>
</select>
```

3. **Estructura jerárquica de títulos:**
```html
<h1>Página principal</h1>
  <h2>Sección 1</h2>
    <h3>Subsección 1.1</h3>
  <h2>Sección 2</h2>
```

4. **ARIA labels cuando sea necesario:**
```html
<button aria-label="Cerrar ventana">×</button>
```

## 2. CSS: Diseño y Presentación

### ¿Qué es CSS?

**CSS** (Cascading Style Sheets) es el lenguaje que define el **aspecto visual** de tu HTML: colores, tamaños, posiciones, animaciones.

### Tres formas de aplicar CSS

**1. CSS Inline (NO recomendado):**
```html
<p style="color: red; font-size: 20px;">Texto rojo</p>
```

**2. CSS Interno (solo para pruebas):**
```html
<head>
    <style>
        p {
            color: red;
            font-size: 20px;
        }
    </style>
</head>
```

**3. CSS Externo (✅ RECOMENDADO):**
```html
<head>
    <link rel="stylesheet" href="estilos.css">
</head>
```
```css
/* estilos.css */
p {
    color: red;
    font-size: 20px;
}
```

### Anatomía de una regla CSS
```css
selector {
    propiedad: valor;
    otra-propiedad: otro-valor;
}
```

Ejemplo:
```css
h1 {
    color: #2C3E50;
    font-size: 32px;
    text-align: center;
}
```

### Selectores CSS: Cómo "apuntar" a elementos

**1. Selector de etiqueta:**
```css
p {
    color: blue;
}
/* Afecta a TODOS los <p> */
```

**2. Selector de clase (`.`):**
```html
<p class="destacado">Texto importante</p>
```
```css
.destacado {
    font-weight: bold;
    background-color: yellow;
}
```

**3. Selector de ID (`#`):**
```html
<div id="resultados">...</div>
```
```css
#resultados {
    width: 100%;
    max-width: 1200px;
}
```

**4. Selectores combinados:**
```css
/* Párrafos dentro de un article */
article p {
    line-height: 1.6;
}

/* Elementos con dos clases */
.candidato.ganador {
    border: 3px solid gold;
}

/* Primer hijo */
li:first-child {
    font-weight: bold;
}

/* Hover (al pasar el mouse) */
a:hover {
    color: red;
}
```

### El Box Model: La caja CSS

Cada elemento HTML es una "caja" con estas capas:
```
┌─────────────────────────────────────┐
│           MARGIN (transparente)     │
│  ┌───────────────────────────────┐  │
│  │    BORDER (borde visible)     │  │
│  │  ┌─────────────────────────┐  │  │
│  │  │  PADDING (relleno)      │  │  │
│  │  │  ┌───────────────────┐  │  │  │
│  │  │  │    CONTENT        │  │  │  │
│  │  │  │   (contenido)     │  │  │  │
│  │  │  └───────────────────┘  │  │  │
│  │  └─────────────────────────┘  │  │
│  └───────────────────────────────┘  │
└─────────────────────────────────────┘
```

Ejemplo:
```css
.card {
    width: 300px;           /* Ancho del contenido */
    padding: 20px;          /* Espacio interno */
    border: 2px solid #ccc; /* Borde */
    margin: 10px;           /* Espacio externo */
}

/* Ancho total = 300 + 20*2 + 2*2 + 10*2 = 364px */
```

**Tip:** Usa `box-sizing: border-box;` para incluir padding y border en el width:
```css
* {
    box-sizing: border-box;
}

.card {
    width: 300px;  /* Ahora el ancho total ES 300px */
    padding: 20px;
    border: 2px solid #ccc;
}
```

### Colores en CSS
```css
/* Nombres de colores */
color: red;

/* Hexadecimal */
color: #FF5733;

/* RGB */
color: rgb(255, 87, 51);

/* RGBA (con transparencia) */
color: rgba(255, 87, 51, 0.5);

/* HSL */
color: hsl(9, 100%, 60%);
```

### Tipografía
```css
body {
    font-family: 'Arial', sans-serif;
    font-size: 16px;
    font-weight: 400;      /* 400 = normal, 700 = bold */
    line-height: 1.6;      /* Altura de línea */
    letter-spacing: 0.5px; /* Espaciado entre letras */
}

h1 {
    font-family: 'Georgia', serif;
    font-size: 2.5rem;     /* rem = relativo al root */
    font-weight: 700;
    text-transform: uppercase;
}
```

### Flexbox: Layout moderno y flexible

Flexbox permite **alinear elementos horizontalmente o verticalmente** de forma sencilla.

**Conceptos clave:**
- **Contenedor flex:** El padre que tiene `display: flex;`
- **Items flex:** Los hijos directos del contenedor
```html
<div class="contenedor">
    <div class="item">1</div>
    <div class="item">2</div>
    <div class="item">3</div>
</div>
```
```css
.contenedor {
    display: flex;
    justify-content: space-between; /* Distribuye horizontalmente */
    align-items: center;            /* Centra verticalmente */
    gap: 20px;                      /* Espacio entre items */
}

.item {
    flex: 1; /* Cada item ocupa el mismo espacio */
}
```

**Propiedades principales del contenedor:**
```css
.flex-container {
    display: flex;
    
    /* Dirección */
    flex-direction: row;        /* horizontal (default) */
    flex-direction: column;     /* vertical */
    
    /* Alineación horizontal (eje principal) */
    justify-content: flex-start;   /* inicio */
    justify-content: center;       /* centro */
    justify-content: space-between; /* espacio entre */
    justify-content: space-around;  /* espacio alrededor */
    
    /* Alineación vertical (eje cruzado) */
    align-items: flex-start; /* arriba */
    align-items: center;     /* centro */
    align-items: flex-end;   /* abajo */
    
    /* Envolver items */
    flex-wrap: wrap; /* Salta a la siguiente línea si no caben */
}
```

**Ejemplo práctico: Navbar con Flexbox**
```html
<nav class="navbar">
    <div class="logo">MiApp Electoral</div>
    <ul class="menu">
        <li><a href="#home">Inicio</a></li>
        <li><a href="#results">Resultados</a></li>
        <li><a href="#stats">Estadísticas</a></li>
    </ul>
</nav>
```
```css
.navbar {
    display: flex;
    justify-content: space-between;
    align-items: center;
    padding: 1rem 2rem;
    background-color: #2C3E50;
    color: white;
}

.menu {
    display: flex;
    gap: 2rem;
    list-style: none;
}

.menu a {
    color: white;
    text-decoration: none;
}
```

### Grid: Layout bidimensional

Grid es perfecto para **layouts complejos** con filas y columnas.
```html
<div class="grid-container">
    <div class="header">Header</div>
    <div class="sidebar">Sidebar</div>
    <div class="main">Main Content</div>
    <div class="footer">Footer</div>
</div>
```
```css
.grid-container {
    display: grid;
    grid-template-columns: 250px 1fr; /* Sidebar fijo, main flexible */
    grid-template-rows: auto 1fr auto;
    grid-template-areas:
        "header header"
        "sidebar main"
        "footer footer";
    gap: 20px;
    min-height: 100vh;
}

.header  { grid-area: header; }
.sidebar { grid-area: sidebar; }
.main    { grid-area: main; }
.footer  { grid-area: footer; }
```

**Ejemplo: Galería de cards**
```css
.galeria {
    display: grid;
    grid-template-columns: repeat(auto-fit, minmax(300px, 1fr));
    gap: 2rem;
}

/* Crea columnas que se ajustan automáticamente, 
   mínimo 300px, máximo lo que quede disponible */
```

### Responsive Design: Media Queries

Las **media queries** permiten aplicar estilos diferentes según el tamaño de pantalla.
```css
/* Estilos base (mobile-first) */
.container {
    width: 100%;
    padding: 1rem;
}

/* Tablets (768px en adelante) */
@media (min-width: 768px) {
    .container {
        width: 750px;
        margin: 0 auto;
    }
}

/* Desktop (1024px en adelante) */
@media (min-width: 1024px) {
    .container {
        width: 960px;
    }
    
    .menu {
        display: flex; /* Menú horizontal en desktop */
    }
}
```

**Breakpoints comunes:**
- Mobile: < 768px
- Tablet: 768px - 1023px
- Desktop: ≥ 1024px

## 3. Bootstrap: Framework CSS

### ¿Qué es Bootstrap?

Bootstrap es un **framework CSS** que proporciona componentes pre-diseñados y un sistema de grillas responsive, ahorrándote tiempo de desarrollo.

### Instalación

**Opción 1: CDN (más fácil para empezar)**
```html
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Mi App con Bootstrap</title>
    
    <!-- Bootstrap CSS -->
    <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/css/bootstrap.min.css" rel="stylesheet">
</head>
<body>
    <!-- Tu contenido -->
    
    <!-- Bootstrap JS (al final del body) -->
    <script src="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/js/bootstrap.bundle.min.js"></script>
</body>
</html>
```

### Sistema de Grillas de Bootstrap

Bootstrap divide la pantalla en **12 columnas**. Vos decidís cuántas columnas ocupa cada elemento.
```html
<div class="container">
    <div class="row">
        <div class="col-md-4">
            <!-- Ocupa 4 columnas (33%) en tablets+ -->
        </div>
        <div class="col-md-8">
            <!-- Ocupa 8 columnas (67%) en tablets+ -->
        </div>
    </div>
</div>
```

**Prefijos de tamaño:**
- `col-` : Todos los tamaños
- `col-sm-` : ≥ 576px (móviles grandes)
- `col-md-` : ≥ 768px (tablets)
- `col-lg-` : ≥ 992px (desktop)
- `col-xl-` : ≥ 1200px (desktop grande)

**Ejemplo: Layout responsive**
```html
<div class="container">
    <div class="row">
        <!-- En móvil: 100% ancho. En tablet+: 33% cada uno -->
        <div class="col-12 col-md-4">
            <div class="card">Candidato 1</div>
        </div>
        <div class="col-12 col-md-4">
            <div class="card">Candidato 2</div>
        </div>
        <div class="col-12 col-md-4">
            <div class="card">Candidato 3</div>
        </div>
    </div>
</div>
```

## 4. JavaScript Básico: Interactividad

### ¿Qué es JavaScript?

JavaScript es el **lenguaje de programación** del navegador. Mientras HTML estructura y CSS estiliza, JavaScript **agrega comportamiento e interactividad**.

### Tres formas de incluir JavaScript

**1. JavaScript Inline (NO recomendado):**
```html
<button onclick="alert('Hola')">Click</button>
```

**2. JavaScript Interno:**
```html
<script>
    console.log("Hola desde JavaScript");
</script>
```

**3. JavaScript Externo (✅ RECOMENDADO):**
```html
<body>
    <!-- Tu HTML -->
    
    <script src="script.js"></script>
</body>
```

⚠️ **Importante:** Coloca el `<script>` al **final del `<body>`** para que el HTML se cargue primero.

### Variables y Tipos de Datos

### Eventos: Reaccionar a la Interacción del Usuario

Los **eventos** permiten ejecutar código cuando algo sucede (click, hover, submit, etc.).

#### addEventListener
```javascript
const boton = document.getElementById("miBoton");

boton.addEventListener("click", function() {
    alert("¡Botón clickeado!");
});

// Con arrow function
boton.addEventListener("click", () => {
    console.log("Click registrado");
});
```

#### Eventos comunes
```javascript
// Click
boton.addEventListener("click", () => {
    console.log("Click");
});

// Doble click
boton.addEventListener("dblclick", () => {
    console.log("Doble click");
});

// Mouse sobre elemento
div.addEventListener("mouseenter", () => {
    console.log("Mouse encima");
});

// Mouse sale del elemento
div.addEventListener("mouseleave", () => {
    console.log("Mouse salió");
});

// Input en formulario
input.addEventListener("input", (e) => {
    console.log("Valor actual:", e.target.value);
});

// Submit de formulario
form.addEventListener("submit", (e) => {
    e.preventDefault(); // Evita el envío tradicional
    console.log("Formulario enviado");
});

// Cambio en select
select.addEventListener("change", (e) => {
    console.log("Opción seleccionada:", e.target.value);
});

// Tecla presionada
document.addEventListener("keydown", (e) => {
    console.log("Tecla presionada:", e.key);
});
```

#### Event Object

El objeto `event` (o `e`) contiene información sobre el evento:
```javascript
boton.addEventListener("click", (e) => {
    console.log(e.target);      // El elemento clickeado
    console.log(e.clientX);     // Posición X del mouse
    console.log(e.clientY);     // Posición Y del mouse
    console.log(e.type);        // Tipo de evento ("click")
});

input.addEventListener("keydown", (e) => {
    console.log(e.key);         // Tecla presionada
    console.log(e.code);        // Código de la tecla
    
    if (e.key === "Enter") {
        console.log("Enter presionado");
    }
});
```

```javascript
// Datos de ejemplo (en la realidad vendrían de una API)
const resultados = [
    { candidato: "Javier Milei", partido: "LLA", provincia: "CABA", votos: 850000, porcentaje: 45.2 },
    { candidato: "Sergio Massa", partido: "UxP", provincia: "CABA", votos: 620000, porcentaje: 32.9 },
    { candidato: "Patricia Bullrich", partido: "JxC", provincia: "CABA", votos: 410000, porcentaje: 21.9 },
    { candidato: "Javier Milei", partido: "LLA", provincia: "Buenos Aires", votos: 2100000, porcentaje: 28.5 },
    { candidato: "Sergio Massa", partido: "UxP", provincia: "Buenos Aires", votos: 3200000, porcentaje: 43.4 },
    { candidato: "Patricia Bullrich", partido: "JxC", provincia: "Buenos Aires", votos: 2100000, porcentaje: 28.1 },
];

// Función para crear una card HTML
function crearCard(resultado) {
    return `
        <div class="col-md-4 mb-4 resultado-item" data-provincia="${resultado.provincia}">
            <div class="card card-resultado">
                <div class="card-body">
                    <h5 class="card-title">${resultado.candidato}</h5>
                    <p class="card-text">
                        <strong>${resultado.partido}</strong><br>
                        ${resultado.provincia}
                    </p>
                    <div class="d-flex justify-content-between align-items-center">
                        <span class="badge bg-primary badge-votos">${resultado.porcentaje}%</span>
                        <small class="text-muted">${resultado.votos.toLocaleString('es-AR')} votos</small>
                    </div>
                </div>
            </div>
        </div>
    `;
}

// Función para renderizar todos los resultados
function renderizarResultados(datos) {
    const contenedor = document.getElementById("resultados");
    contenedor.innerHTML = datos.map(crearCard).join("");
}

// Función para filtrar resultados
function filtrarPorProvincia(provincia) {
    const items = document.querySelectorAll(".resultado-item");
    
    items.forEach(item => {
        if (provincia === "todas" || item.dataset.provincia === provincia) {
            item.style.display = "block";
        } else {
            item.style.display = "none";
        }
    });
}

// Event listener para el select
document.getElementById("provinciaSelect").addEventListener("change", (e) => {
    filtrarPorProvincia(e.target.value);
});

// Renderizar resultados al cargar la página
renderizarResultados(resultados);
```

## 5. Responsive Design y Mobile-First

### ¿Qué es Responsive Design?

Diseño que **se adapta** a diferentes tamaños de pantalla: móviles, tablets, desktop.

### Enfoque Mobile-First

Diseñar primero para móviles, luego agregar funcionalidades para pantallas más grandes.
```css
/* Base: Móvil (< 768px) */
.container {
    padding: 1rem;
    font-size: 14px;
}

/* Tablet (≥ 768px) */
@media (min-width: 768px) {
    .container {
        padding: 2rem;
        font-size: 16px;
    }
}

/* Desktop (≥ 1024px) */
@media (min-width: 1024px) {
    .container {
        max-width: 1200px;
        margin: 0 auto;
        font-size: 18px;
    }
}
```

### Viewport Meta Tag

**OBLIGATORIO** para responsive design:
```html
<meta name="viewport" content="width=device-width, initial-scale=1.0">
```

### Unidades Responsive
```css
/* NO usar px fijos para todo */
.malo {
    width: 500px; /* Se ve mal en móviles pequeños */
}

/* Usar unidades relativas */
.bueno {
    width: 90%;           /* Porcentaje del padre */
    max-width: 1200px;    /* Límite máximo */
    padding: 2rem;        /* rem = relativo al root (16px por defecto) */
    font-size: 1.5rem;    /* 1.5 * 16px = 24px */
}
```

### Imágenes Responsive
```css
img {
    max-width: 100%;
    height: auto;
}
```
```html
<!-- Picture element para diferentes resoluciones -->
<picture>
    <source media="(max-width: 768px)" srcset="imagen-mobile.jpg">
    <source media="(max-width: 1200px)" srcset="imagen-tablet.jpg">
    <img src="imagen-desktop.jpg" alt="Descripción">
</picture>
```

```javascript
// Datos de ejemplo (en producción vendrían de una API)
const datosElectorales = {
    mesasEscrutadas: 87.5,
    totalVotos: 18542000,
    ultimaActualizacion: new Date().toLocaleString('es-AR'),
    candidatos: [
        {
            nombre: "Javier Milei",
            partido: "La Libertad Avanza",
            alias: "LLA",
            votos: 5650000,
            porcentaje: 30.5,
            color: "partido-lla"
        },
        {
            nombre: "Sergio Massa",
            partido: "Unión por la Patria",
            alias: "UxP",
            votos: 5340000,
            porcentaje: 28.8,
            color: "partido-uxp"
        },
        {
            nombre: "Patricia Bullrich",
            partido: "Juntos por el Cambio",
            alias: "JxC",
            votos: 4120000,
            porcentaje: 22.2,
            color: "partido-jxc"
        }
    ]
};

// Función para actualizar estadísticas generales
function actualizarEstadisticas() {
    document.getElementById('mesasEscrutadas').textContent = 
        datosElectorales.mesasEscrutadas + '%';
    
    document.getElementById('totalVotos').textContent = 
        datosElectorales.totalVotos.toLocaleString('es-AR');
    
    document.getElementById('ultimaActualizacion').textContent = 
        datosElectorales.ultimaActualizacion;
}

// Función para crear card de candidato
function crearCardCandidato(candidato) {
    return `
        <div class="col-md-4 candidato-item fade-in" data-partido="${candidato.alias}">
            <div class="card card-candidato">
                <div class="card-header ${candidato.color}">
                    ${candidato.nombre}
                </div>
                <div class="card-body">
                    <h5 class="card-title">${candidato.partido}</h5>
                    <div class="mb-3">
                        <span class="badge ${candidato.color} badge-porcentaje">
                            ${candidato.porcentaje}%
                        </span>
                    </div>
                    <div class="progress mb-3">
                        <div class="progress-bar ${candidato.color} progress-bar-custom" 
                             role="progressbar" 
                             style="width: ${candidato.porcentaje}%"
                             aria-valuenow="${candidato.porcentaje}" 
                             aria-valuemin="0" 
                             aria-valuemax="100">
                            ${candidato.porcentaje}%
                        </div>
                    </div>
                    <p class="card-text">
                        <strong>Votos:</strong> ${candidato.votos.toLocaleString('es-AR')}<br>
                        <small class="text-muted">Representan el ${candidato.porcentaje}% del total</small>
                    </p>
                </div>
            </div>
        </div>
    `;
}

// Función para renderizar todos los candidatos
function renderizarResultados(candidatos) {
    const contenedor = document.getElementById('contenedorResultados');
    contenedor.innerHTML = candidatos.map(crearCardCandidato).join('');
}

// Función para filtrar por búsqueda de candidato
function filtrarPorNombre(termino) {
    const items = document.querySelectorAll('.candidato-item');
    const terminoLower = termino.toLowerCase();
    
    items.forEach(item => {
        const texto = item.textContent.toLowerCase();
        if (texto.includes(terminoLower)) {
            item.style.display = 'block';
        } else {
            item.style.display = 'none';
        }
    });
}

// Event Listeners
document.addEventListener('DOMContentLoaded', () => {
    // Inicializar datos
    actualizarEstadisticas();
    renderizarResultados(datosElectorales.candidatos);
    
    // Búsqueda por nombre
    document.getElementById('busquedaCandidato').addEventListener('input', (e) => {
        filtrarPorNombre(e.target.value);
    });
    
    // Simulador de actualización en tiempo real (cada 5 segundos)
    setInterval(() => {
        datosElectorales.mesasEscrutadas = Math.min(100, datosElectorales.mesasEscrutadas + 0.5);
        datosElectorales.ultimaActualizacion = new Date().toLocaleString('es-AR');
        actualizarEstadisticas();
    }, 5000);
});

// Mostrar mensaje de bienvenida
console.log('%c🗳️ Dashboard Electoral 2023', 'font-size: 20px; font-weight: bold; color: #2C3E50;');
console.log('%cDatos provisorios - Actualización automática cada 5 segundos', 'color: #7F8C8D;');
```

## 9. Resumen y Conceptos Clave

### Checklist de Conocimientos

Después de esta clase deberías poder:

- [ ] Estructurar una página web con HTML5 semántico
- [ ] Aplicar estilos con CSS (colores, tipografía, layout)
- [ ] Usar Flexbox y Grid para layouts
- [ ] Implementar diseño responsive con media queries
- [ ] Usar componentes de Bootstrap (navbar, cards, forms, modales)
- [ ] Manipular el DOM con JavaScript
- [ ] Manejar eventos (click, input, submit)
- [ ] Crear, modificar y eliminar elementos dinámicamente
- [ ] Integrar HTML + CSS + JavaScript en un proyecto funcional

### Conceptos Fundamentales

**HTML:**
- Estructura semántica (`<header>`, `<main>`, `<article>`, etc.)
- Accesibilidad (atributo `alt`, `<label>`, estructura jerárquica)

**CSS:**
- Box Model (content, padding, border, margin)
- Selectores (etiqueta, clase, ID, combinados)
- Flexbox (para layouts en una dimensión)
- Grid (para layouts bidimensionales)
- Media Queries (diseño responsive)

**JavaScript:**
- Variables (`let`, `const`)
- Funciones (clásicas y arrow functions)
- DOM Manipulation (`querySelector`, `getElementById`)
- Event Listeners (`addEventListener`)
- Crear/modificar/eliminar elementos

**Bootstrap:**
- Sistema de grillas (12 columnas)
- Componentes pre-diseñados (navbar, cards, modales)
- Clases de utilidad (espaciado, colores, texto)

**Objetivo:** Crear un "Dashboard de Datos Políticos" que consuma datos de una API simulada.

**Requisitos:**
1. HTML semántico con estructura clara
2. Navbar con Bootstrap
3. Al menos 3 secciones diferentes
4. Tabla de datos con Bootstrap (table-striped, table-hover)
5. Gráfico simple con colores (pueden usar divs con background-color por ahora)

7. Input de búsqueda que filtre resultados en tiempo real
8. Diseño 100% responsive

**Bonus:**
- Implementar modo oscuro con toggle
- Agregar animaciones con CSS
